```python
import requests
from bs4 import BeautifulSoup
import numpy as np

url = 'https://wonder.cdc.gov/nndss/static/2019/annual/2019-table2l.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html")
print(soup.prettify)
```

```python
# Need to web scrape multiple charts 

# for 2019

def webscrape(soup):
    # state placeholders for iteration
    states_placehold = [4,5,6,7,8,9,11,12,13,14,16,17,18,19,20,22,23,24,25,26,27,28,30,31,32,33,34,35,36,37,38,40,41,42,43,45,46,47,48,50,51,52,53,54,55,56,57,59,60,61,62,63]

    # empty lists to append
    states_list = []
    mumps_cases = []

    div_class = soup.find('div', class_ = "container d-flex flex-wrap body-wrapper")
    main_class = div_class.find('main')
    div__row = main_class.find_all('div', class_ = "row")
    div_col_content = div__row[2].find('div', class_ = "col content")
    div_nndss = div_col_content.find('div', class_ = "nndss-content")
    table_nndss = div_nndss.find('table')
    tr = table_nndss.find_all('tr')
    for state in states_placehold:
        th = tr[state].find('th')
        td = tr[state].find_all('td')
        states_list.append(th.find(text=True))
        mumps_cases.append(td[5].find(text=True))
        
    return states_list, mumps_cases

webscrape(soup)

states_list, mumps_cases = webscrape(soup)

mumps_2019 = pd.DataFrame({'state':states_list,
                        'cum_mumps_2019':mumps_cases
                               })
mumps_2019.head(10)
```

```python
# For 2011 to 2015, and 2020 annual reports were not made but there are weekly reports and the week 52 reports that contains 
# the same data for cum_mumps (cumulative number of mumps cases for particular year) as the annual.
# I noticed that the a column for the previous year's cumulative mumps cases is also in the proceeding years report and more
# accurately represents the cumulatives mumps count found in the annual reports.  Having worked in a lab in previous 
# experiences, I can say that there may have been a delay in transit or suspicion of contamination, which may have resulted in
# the need for a redraw.  In which case I am using the cumulative results found in the proceeding years reports where the
# cumulative results have been acturately updated.  However, I will need to use to use the 2020 report for 2020, as 2021 is only
# on week 39.

# for 2020

url = 'https://wonder.cdc.gov/nndss/static/2020/52/2020-52-table1y.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html")

# state placeholders for iteration
states_placehold = [4,5,6,7,8,9,11,12,13,14,16,17,18,19,20,22,23,24,25,26,27,28,30,31,32,33,34,35,36,37,38,40,41,42,43,45,46,47,48,50,51,52,53,54,55,56,57,59,60,61,62,63]

# empty lists to append
states_list = []
mumps_cases = []

div_class = soup.find('div', class_ = "container d-flex flex-wrap body-wrapper")
main_class = div_class.find('main')
div__row = main_class.find_all('div', class_ = "row")
div_col_content = div__row[2].find('div', class_ = "col content")
div_nndss = div_col_content.find('div', class_ = "nndss-content")
table_nndss = div_nndss.find('table')
tr = table_nndss.find_all('tr')
for state in states_placehold:
    th = tr[state].find('th')
    td = tr[state].find_all('td')
    states_list.append(th.find(text=True))
    mumps_cases.append(td[2].find(text=True))

mumps_2020 = pd.DataFrame({'state':states_list,
                        'cum_mumps_2020':mumps_cases
                               })
mumps_2020.head(10)
```

```python
# At this point I did realize that the table can be exported but I already have functions writen to webscrap and it actually
# saves some work for data wrangling.

# for 2015

url = 'https://wonder.cdc.gov/nndss/nndss_weekly_tables_2016_38.asp?mmwr_year=2016&mmwr_week=52&mmwr_table=2I&request=Submit'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html")

def webscrape2(soup):
    # state placeholders for iteration
    states_placehold = [5,6,7,8,9,10,12,13,14,15,17,18,19,20,21,23,24,25,26,27,28,29,31,32,33,34,35,36,37,38,39,41,42,43,44,46,47,48,49,51,52,53,54,55,56,57,58,60,61,62,63,64]
    
    # empty lists to append
    states_list = []
    mumps_cases = []

    div_class = soup.find('div', class_ = "container d-flex flex-wrap body-wrapper")
    main_class = div_class.find('main')
    div__row = main_class.find_all('div', class_ = "row")
    div_col_content = div__row[2].find('div', class_ = "col content")
    div_nndss = div_col_content.find('div', class_ = "nndss-content")
    table_nndss = div_nndss.find('table')
    tr = table_nndss.find_all('tr')
    for state in states_placehold:
        th = tr[state].find('th')
        td = tr[state].find_all('td')
        states_list.append(th.find(text=True))
        mumps_cases.append(td[4].find(text=True))
        
    return states_list, mumps_cases

webscrape2(soup)

states_list, mumps_cases = webscrape2(soup)    

mumps_2015 = pd.DataFrame({'state':states_list,
                        'cum_mumps_2015':mumps_cases
                               })
mumps_2015.head(10)
```

```python
# for 2014

url = 'https://wonder.cdc.gov/nndss/nndss_weekly_tables_2015.asp?mmwr_year=2015&mmwr_week=52&mmwr_table=2I&request=Submit'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html")

def webscrape3(soup):
    # state placeholders for iteration
    states_placehold = [5,6,7,8,9,10,12,13,14,15,17,18,19,20,21,23,24,25,26,27,28,29,31,32,33,34,35,36,37,38,39,41,42,43,44,46,47,48,49,51,52,53,54,55,56,57,58,60,61,62,63,64]
    
    # empty lists to append
    states_list = []
    mumps_cases = []

    div_class = soup.find('div', class_ = "container d-flex flex-wrap body-wrapper")
    main_class = div_class.find('main')
    div__row = main_class.find_all('div', class_ = "row")
    div_col_content = div__row[2].find('div', class_ = "col content")
    div_mmwr = div_col_content.find('div', class_ = "mmwr-content")
    table_mmwr = div_mmwr.find('table')
    tr = table_mmwr.find_all('tr')
    for state in states_placehold:
        td = tr[state+1].find_all('td')
        states_list.append(td[0].find(text=True))
        mumps_cases.append(td[5].find(text=True))
        
    return states_list, mumps_cases

webscrape3(soup)

states_list, mumps_cases = webscrape3(soup)    

mumps_2014 = pd.DataFrame({'state':states_list,
                        'cum_mumps_2014':mumps_cases
                               })
mumps_2014.head(10)
```
